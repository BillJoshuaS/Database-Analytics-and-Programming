```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load your data into a pandas DataFrame
df = pd.read_csv('Unempl_US_State.csv')

# Calculate average employment rate for each year
average_employment_by_year = df.groupby('Year')['Percent (%) of Labor Force Employed in State/Area'].mean().reset_index()


# Plotting the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_employment_by_year, x='Year', y='Percent (%) of Labor Force Employed in State/Area', marker='o', markerfacecolor='b', c = 'g', alpha = 0.5)
plt.title('Average Employment Rate Over 46 Years')
plt.xlabel('Year', fontsize=16, fontweight='bold', fontfamily='sans-serif')
plt.ylabel('Average Employment Rate (%)', fontsize=16, fontweight='bold', fontfamily='sans-serif')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

average_unemployment_by_year = df.groupby('Year')['Percent (%) of Labor Force Unemployed in State/Area'].mean().reset_index()


# Plotting the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_unemployment_by_year, x='Year', y='Percent (%) of Labor Force Unemployed in State/Area', marker='o', c = 'r', markerfacecolor='b', alpha = 0.5)
plt.title('Average Unemployment Rate Over 46 Years')
plt.xlabel('Year', fontsize=16, fontweight='bold', fontfamily='sans-serif')
plt.ylabel('Average Unemployment Rate (%)', fontsize=10, fontweight='bold', fontfamily='sans-serif')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()


```

```python

# Load the data
df = pd.read_csv("Unemployment.csv")

# Group by state and calculate the sum of total income and total cost
df_state_totals = df.groupby('State/Area')[['Total Employment in State/Area', 'Total Unemployment in State/Area']].sum().reset_index()

# Plotting with seaborn lineplot
plt.figure(figsize=(14, 8))

# Set the seaborn style to 'ticks' and specify color palette
sns.set_style("whitegrid")
colors = sns.color_palette("bright", 2)
sns.lineplot(data=df_state_totals, x='State/Area', y='Total Employment in State/Area', label='Total Employment', marker='D', color=colors[0])
sns.lineplot(data=df_state_totals, x='State/Area', y='Total Unemployment in State/Area', label='Total Unemployment', marker='D', color=colors[1])

# Adding labels and title with adjusted font size
plt.title('Total Employment and Total Unemployment by State', fontsize=18, fontweight='bold', fontfamily='sans-serif')
plt.xlabel('State/Area', fontsize=14, fontweight='bold', fontfamily='sans-serif')
plt.ylabel('Employment and Unemployment', fontsize=14, fontweight='bold', fontfamily='sans-serif')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Set y-axis range to ensure all data points are visible
plt.ylim(0, df_state_totals[['Total Employment in State/Area', 'Total Unemployment in State/Area']].values.max() * 1.1)

# Adjust legend placement to avoid overlapping with the plot
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("Unemployment.csv")

X = df[['Total Civilian Non-Institutional Population in State/Area','Total Civilian Labor Force in State/Area','Total Unemployment in State/Area']]
y = df['Total Employment in State/Area']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=52)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

rf_regressor = RandomForestRegressor(n_estimators=100, max_depth=12, random_state=0)

rf_regressor.fit(X_train_scaled, y_train)

y_pred_train_rf = rf_regressor.predict(X_train_scaled)
y_pred_test_rf = rf_regressor.predict(X_test_scaled)

r2_train_rf = r2_score(y_train, y_pred_train_rf)
r2_test_rf = r2_score(y_test, y_pred_test_rf)

print('Training data R2 score for Employment rate prediction: ', r2_train_rf)
print('Testing data R2 score for Employment rate prediction: ', r2_test_rf)

plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_pred_train_rf, color='blue', label='Training data', alpha=0.5)
plt.scatter(y_test, y_pred_test_rf, color='red', label='Testing data', alpha=0.5)
plt.title('Actual vs Predicted Values for Random Forest Regressor')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
